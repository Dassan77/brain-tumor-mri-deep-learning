```python
# ============================================================
# FINAL CLEANED WORKING CODE (NO Addons) - 4-Class Brain Tumor MRI
# Dataset: sartajbhuvaji/brain-tumor-classification-mri
#
# Works on Colab with NO tensorflow-addons.
# Improvements included:
# ✅ EfficientNetB0 backbone
# ✅ Class weights (helps glioma/minority classes)
# ✅ Two-phase training (head + fine-tuning)
# ✅ Confusion matrix + classification report
# ✅ Sample predictions
# ✅ Single-image prediction (upload an image)
# ✅ Saves model as .keras
# ============================================================

# -------------------------
# 0) Install
# -------------------------
!pip -q install kaggle tensorflow matplotlib scikit-learn numpy

# -------------------------
# 1) Imports
# -------------------------
import os
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight

# -------------------------
# 2) Upload kaggle.json (once per session)
# -------------------------
from google.colab import files

kaggle_json_path = os.path.expanduser("~/.kaggle/kaggle.json")
if not os.path.exists(kaggle_json_path):
    print("Upload kaggle.json now:")
    files.upload()  # select kaggle.json

    os.makedirs(os.path.expanduser("~/.kaggle"), exist_ok=True)
    !cp kaggle.json ~/.kaggle/
    !chmod 600 ~/.kaggle/kaggle.json
else:
    print("kaggle.json already exists. Skipping upload.")

# -------------------------
# 3) Download dataset
# -------------------------
DATASET_SLUG = "sartajbhuvaji/brain-tumor-classification-mri"

!rm -rf /content/data_4class
!mkdir -p /content/data_4class
!kaggle datasets download -d $DATASET_SLUG -p /content/data_4class --unzip

print("\nUnzipped top-level folders/files:")
print(os.listdir("/content/data_4class"))

# -------------------------
# 4) Auto-find Training/Testing
# -------------------------
def find_first_dir(root, target_name):
    for dirpath, dirnames, filenames in os.walk(root):
        if target_name in dirnames:
            return os.path.join(dirpath, target_name)
    return None

TRAIN_DIR = find_first_dir("/content/data_4class", "Training")
TEST_DIR  = find_first_dir("/content/data_4class", "Testing")

print("\nTRAIN_DIR:", TRAIN_DIR)
print("TEST_DIR :", TEST_DIR)

if TRAIN_DIR is None or TEST_DIR is None:
    raise ValueError("Could not find Training/Testing folders. Inspect /content/data_4class")

print("\nTrain class folders:", os.listdir(TRAIN_DIR))
print("Test class folders :", os.listdir(TEST_DIR))

# -------------------------
# 5) Parameters
# -------------------------
IMG_SIZE = (256, 256)  # if slow, try (224,224) or (192,192)
BATCH_SIZE = 16        # if you have strong GPU, try 32
SEED = 42
AUTOTUNE = tf.data.AUTOTUNE

# -------------------------
# 6) Load datasets (categorical labels)
# -------------------------
train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_DIR,
    labels="inferred",
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    TEST_DIR,
    labels="inferred",
    label_mode="categorical",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
num_classes = len(class_names)
print("\nClasses:", class_names, "num_classes:", num_classes)

train_ds = train_ds.prefetch(AUTOTUNE)
test_ds  = test_ds.prefetch(AUTOTUNE)

# -------------------------
# 7) Train/Val split from training (80/20 by batches)
# -------------------------
train_batches = tf.data.experimental.cardinality(train_ds).numpy()
train_size = int(0.8 * train_batches)

train_ds2 = train_ds.take(train_size).prefetch(AUTOTUNE)
val_ds    = train_ds.skip(train_size).prefetch(AUTOTUNE)

print("\nBatches -> train:", tf.data.experimental.cardinality(train_ds2).numpy(),
      "val:", tf.data.experimental.cardinality(val_ds).numpy())

# -------------------------
# 8) Compute class weights (balanced)
# -------------------------
y_train_idx = []
for _, labels in train_ds2:
    y_train_idx.extend(np.argmax(labels.numpy(), axis=1).tolist())
y_train_idx = np.array(y_train_idx)

weights = compute_class_weight(
    class_weight="balanced",
    classes=np.arange(num_classes),
    y=y_train_idx
)
class_weight = {i: float(w) for i, w in enumerate(weights)}
print("\nComputed class_weight:", class_weight)

# -------------------------
# 9) Augmentation + preprocessing (EfficientNet)
# -------------------------
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.12),
    layers.RandomZoom(0.15),
    layers.RandomContrast(0.15),
], name="augmentation")

preprocess_input = keras.applications.efficientnet.preprocess_input

# -------------------------
# 10) Build model (EfficientNetB0 Transfer Learning)
# -------------------------
base_model = keras.applications.EfficientNetB0(
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3),
    include_top=False,
    weights="imagenet"
)
base_model.trainable = False

inputs = keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.35)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("\nMODEL SUMMARY:")
model.summary()

callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6),
]

# -------------------------
# 11) Train Phase 1 (head training)
# -------------------------
history1 = model.fit(
    train_ds2,
    validation_data=val_ds,
    epochs=18,
    callbacks=callbacks,
    class_weight=class_weight,
    verbose=1
)

# -------------------------
# 12) Fine-tune Phase 2
# -------------------------
base_model.trainable = True

fine_tune_at = 80  # lower -> more layers trainable (better but slower)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history2 = model.fit(
    train_ds2,
    validation_data=val_ds,
    epochs=12,
    callbacks=callbacks,
    class_weight=class_weight,
    verbose=1
)

# -------------------------
# 13) Plot curves
# -------------------------
def plot_history(h, title):
    plt.figure()
    plt.plot(h.history["loss"], label="train_loss")
    plt.plot(h.history["val_loss"], label="val_loss")
    plt.title(title + " Loss")
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(h.history["accuracy"], label="train_acc")
    plt.plot(h.history["val_accuracy"], label="val_acc")
    plt.title(title + " Accuracy")
    plt.legend()
    plt.show()

plot_history(history1, "Phase 1")
plot_history(history2, "Phase 2")

# -------------------------
# 14) Evaluate on TEST
# -------------------------
test_loss, test_acc = model.evaluate(test_ds, verbose=1)
print("\n===== TEST RESULTS =====")
print("Test Loss:", test_loss)
print("Test Accuracy:", test_acc)

# -------------------------
# 15) Confusion matrix + classification report
# -------------------------
y_true, y_pred = [], []

for images, labels in test_ds:
    probs = model.predict(images, verbose=0)
    y_true.extend(np.argmax(labels.numpy(), axis=1).tolist())
    y_pred.extend(np.argmax(probs, axis=1).tolist())

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:\n", cm)

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))

# -------------------------
# 16) Show sample predictions (for presentation)
# -------------------------
plt.figure(figsize=(12, 8))
images, labels = next(iter(test_ds))
probs = model.predict(images, verbose=0)

for i in range(min(9, images.shape[0])):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    t = class_names[int(np.argmax(labels[i].numpy()))]
    p = class_names[int(np.argmax(probs[i]))]
    plt.title(f"T:{t}\nP:{p}", fontsize=9)
    plt.axis("off")

plt.tight_layout()
plt.show()

# -------------------------
# 17) Save model
# -------------------------
model.save("brain_tumor_4class_efficientnet_best.keras")
print("\nSaved: brain_tumor_4class_efficientnet_best.keras")

# -------------------------
# 18) Predict ONE image (upload)
# -------------------------
from tensorflow.keras.preprocessing import image

def predict_single_image(img_path):
    img = image.load_img(img_path, target_size=IMG_SIZE)
    arr = image.img_to_array(img)
    arr = np.expand_dims(arr, axis=0)
    arr = preprocess_input(arr)

    probs = model.predict(arr, verbose=0)[0]
    pred_idx = int(np.argmax(probs))
    pred_class = class_names[pred_idx]
    conf = float(probs[pred_idx])

    plt.figure()
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Prediction: {pred_class} | Confidence: {conf:.2f}")
    plt.show()

    return pred_class, conf

print("\nUpload ONE MRI image to test:")
up = files.upload()
img_path = list(up.keys())[0]
predict_single_image(img_path)
```
